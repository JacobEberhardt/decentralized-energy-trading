```python
import pandas as pd
import glob
import numpy as np
import matplotlib
import json
import datetime
```

```python
# get data from current folder with json files
path = r'.' 
all_files = glob.glob("./*.json")
    
df_list = []
for index, js in enumerate(all_files):
    with open(js) as json_file:  
        data = json.load(json_file)
        print("", index, ": File ",js," joined with ", len(data)," entries.")
        single_df = pd.DataFrame(data)
        df_list.append(single_df)
df = pd.concat(df_list,sort=False, ignore_index=True)
```

```python
df['date'] = (pd.to_datetime(df['IEnd'], unit='ms'))
df['year'] = (pd.to_datetime(df['IEnd'], unit='ms')).apply(lambda x: x.strftime('%Y'))
df['month'] = (pd.to_datetime(df['IEnd'], unit='ms')).apply(lambda x: x.strftime('%m'))
df['day'] = (pd.to_datetime(df['IEnd'], unit='ms')).apply(lambda x: x.strftime('%d'))
df['hour'] = (pd.to_datetime(df['IEnd'], unit='ms')).apply(lambda x: x.strftime('%H'))
df['quarterhour'] = (pd.to_datetime(df['IEnd'], unit='ms')).apply(lambda x: x.strftime('%M'))
```
